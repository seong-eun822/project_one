```python
LANL_D=LANL.drop(['Job Number', 'Requested Memory', 'Executable (Application) Number','Used Memory', 'Partition Number', 'Preceding Job Number','Think Time from Preceding Job'], axis=1)
LANL_D
```

```python
LANL_D['Status'].value_counts()
```

```python
LANL_D.drop(LANL_D[LANL_D['Status'] == 0].index, inplace=True)
LANL_D
```

```python
cols=['Wait Time', 'Number of Allocated Processors', 
     'Average CPU Time Used','Requested Number of Processors', 'Requested Time']
```

```python

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.inspection import PartialDependenceDisplay
# ==== 1. 기본 전처리 ====

target='Run Time'
features = [c for c in cols if c in LANL_D.columns]
LANL_D = LANL_D.dropna(subset=features + [target]).copy()

q1, q3 = LANL_D['Run Time'].quantile([0.25, 0.75])
iqr = q3 - q1
LANL_D = LANL_D[(LANL_D['Run Time'] > q1 - 1.5*iqr) & (LANL_D['Run Time'] < q3 + 1.5*iqr)]

# ==== 2. (선택) 로그 파생변수 ====
for c in ['Number of Allocated Processors','Requested Time','Wait Time','Average CPU Time Used','Requested Number of Processors']:
    if c in LANL_D.columns:
        LANL_D[f'log1p_{c}'] = np.log1p(LANL_D[c]) #큰 스케일 변수의 폭주를 막아주는 전처리
y = np.log1p(LANL_D['Run Time']) #y또한 폭주를 막는 전처리

X_cols = [c for c in features if c in LANL_D.columns] + \
         [f'log1p_{c}' for c in features if f'log1p_{c}' in LANL_D.columns]
X = LANL_D[X_cols]
y = LANL_D[target]

# ==== 3. Train/Test 분리 ====
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# ==== 4. 랜덤포레스트 학습 ====
model = RandomForestRegressor(
    n_estimators=400,
    max_depth=None,
    min_samples_leaf=8,
    max_features=0.7,
    n_jobs=-1,
    random_state=42
)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("R² (test):", round(r2_score(y_test, y_pred), 3)) #R²: 모델이 Run Time 변동의 몇%를 설명하는가
print("RMSE (test):", round(np.sqrt(mean_squared_error(y_test, y_pred)), 3)) #예측 오차의 평균 크기

# ==== 5. 변수 중요도 ====
importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=True)
plt.figure(figsize=(6, max(3, 0.35*len(importances))))
importances.plot(kind='barh')
plt.title('Variable Importance (Random Forest)')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()

# ==== 6. 부분의존도(PDP) ====
topk = importances.sort_values(ascending=False).head(3).index.tolist()
PartialDependenceDisplay.from_estimator(model, X, features=topk, kind='average')
plt.suptitle('Partial Dependence (Top Variables)', y=1.02)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import PartialDependenceDisplay
from scipy.optimize import curve_fit

# (1) 모델 훈련
rf = RandomForestRegressor(random_state=42)
rf.fit(X, y)

# (2) PDP 계산
disp = PartialDependenceDisplay.from_estimator(
    rf, X, ['Average CPU Time Used'], grid_resolution=50
)
x_vals = disp.lines_[0][0].get_xdata()
y_vals = disp.lines_[0][0].get_ydata()

# (3) 로그함수 모델 피팅 (y = a·log(1+x) + b)
def log_func(x, a, b):
    return a * np.log1p(x) + b

params, _ = curve_fit(log_func, x_vals, y_vals, maxfev=10000)
a, b = params

# (4) 로그함수 예측선 추가
x_fit = np.linspace(x_vals.min(), x_vals.max(), 300)
y_fit = log_func(x_fit, a, b)

plt.plot(x_fit, y_fit, color='red', linewidth=2, label=f'Log fit: a·log(1+x)+b')
plt.legend()
plt.show()

print(f"📈 로그근사: y = {a:.3f}·log(1+x) + {b:.3f}")

```

```python
#1️⃣ 변수 중요도 그래프 (Variable Importance plot)
#2️⃣ 부분 의존도 그래프 (Partial Dependence Plot, PDP)

#산점도->  선형관계는 안보임 세로로 긴 형태
#단순히 두 변수만 확인 
#하지만 런타임은  다른 변수의 조합에 따라 결정됨.=> 따라서 산점도에서는 다른 변수들의 영향이 섞여있어서 관계가 희미하게 보임.

#부분 의존도 그래프-> 비선형적, 구간별로 다른 형태
#다른 변수들의 평균적 영향을 제거한 "순수한 효과"만 보여줌.
#다른 변수는 고정해두고, 한 변수의 효과만 보여줌.

#랜덤포레스트에서 R²: 모델이 run time의 변동의 몇%를 설명하는가
#0.6 이상은 나와야 좋은 모델이라고 할 수 있음.
```

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import PartialDependenceDisplay
from scipy.optimize import curve_fit

# (1) 모델 훈련
rf = RandomForestRegressor(random_state=42)
rf.fit(X, y)

# (2) PDP 계산
disp = PartialDependenceDisplay.from_estimator(
    rf, X, ['log1p_Average CPU Time Used'], grid_resolution=50
)
x_vals = disp.lines_[0][0].get_xdata()
y_vals = disp.lines_[0][0].get_ydata()

# (3) 지수함수 모델 피팅: y = a * exp(b * x) + c
def exp_func(x, a, b, c):
    return a * np.exp(b * x) + c

params, _ = curve_fit(exp_func, x_vals, y_vals, maxfev=20000)
a, b, c = params

# (4) 지수함수 예측선 추가
x_fit = np.linspace(x_vals.min(), x_vals.max(), 300)
y_fit = exp_func(x_fit, a, b, c)

plt.plot(x_fit, y_fit, linewidth=2,color='#ff7f0e', label=f'Exp fit: a·exp(b·x)+c')
plt.legend()
plt.show()

print(f"📈 지수근사: y = {a:.3f}·exp({b:.3f}·x) + {c:.3f}")
```
